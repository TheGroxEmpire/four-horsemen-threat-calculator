```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
while(True):
    tankTps = int(input("Tank threat per second: "))
    if tankTps >= 0:
        break
    else:
        print("Please insert a valid number")
        
while(True):
    dpsTps = int(input("DPS threat per second: "))
    if dpsTps >= 0:
        break
    else:
        print("Please insert a valid number")

tps = [tankTps, dpsTps]
firstMark = 20
markInterval = 12
timeLimit = 112
sumThreatArray = []
n = 0

for i in tps:
    temp = []
    time = 0
    sumThreat = 0
    n += 1
    while(time < timeLimit):
        time += 1
        sumThreat += i
        if(time == 12 or time > 12 and (time - 12) % 20 == 0):
            sumThreat = sumThreat / 2
            if n == 1:
                print("Tank threat @", time,"s :", sumThreat)
            else:
                print("DPS threat @", time,"s :", sumThreat)
        
        temp.append(sumThreat)
    sumThreatArray.append(temp)
x = range(timeLimit)
y = np.asarray(sumThreatArray)
labels = ['Tank Threat', 'DPS Threat']

plt.figure(figsize=(15,5))
for y_arr, label in zip(y, labels):
    plt.plot(x, y_arr, label=label)
plt.legend()
plt.xlabel('Seconds') 
plt.ylabel('Total Threats') 
plt.title('Four Horsemen Threat Graph')
plt.xticks(np.arange(0, time, 25))
plt.grid(True)
plt.show()
```
